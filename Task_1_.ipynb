## Analysis of Video Views by Resolution

In this analysis, we present the percentage of video views based on video resolution (SD and HD) using a pie chart created with Plotly Express. This visualization aims to illustrate the distribution of video views between the different resolutions and understand user preferences.

### Key Points in the Analysis:
**1. Calculation of the Percentage of Views for Each Resolution**:
- The total number of views for all resolutions was calculated.
- The percentage for each resolution was calculated by dividing the number of views for each resolution by the total number of views and multiplying the result by 100 to obtain the percentage.

**2. Creating a DataFrame for Plotly**:
- A DataFrame was created containing the resolutions (SD and HD) and the percentages of views for each resolution.

**3. Visualizing the Data Using a Pie Chart**:
- The pie chart was created using the Plotly Express library.
- The colors of the chart were customized to be more serious, using green (#2e8b57) for SD resolution and blue (#4682b4) for HD resolution.
- Displaying the percentage and label inside the pie chart slices was enabled for better clarity of information.

```python
from google.colab import files
import pandas as pd
import io

# Upload the file
uploaded = files.upload()

# Iterate over the uploaded files (in case multiple files are uploaded)
for filename in uploaded.keys():
    # Read the first sheet in the Excel file
    dataframe = pd.read_excel(io.BytesIO(uploaded[filename]), sheet_name=0)  # 0 refers to the first sheet
    print(dataframe.head())  # Display the first few rows of the dataframe
```

```python
# check the data shape
dataframe.shape
```

```python
# display the first 5 rows
dataframe.head()
```

```python
# Data Preprocessing on the input data
dataframe = dataframe.drop(columns=['Column1']) # dropping the index column
dataframe['program_name'] = dataframe['program_name'].str.strip()  # trim spaces in movies names to avoid misspellings in input data
dataframe[['duration_seconds', 'season','episode','series_title','hd']] = dataframe[['duration_seconds', 'season','episode','series_title','hd']].apply(pd.to_numeric)  # read numeric columns as numeric data types
dataframe[['user_id_maped', 'program_name','program_class','program_desc','program_genre','original_name']] = dataframe[['user_id_maped', 'program_name','program_class','program_desc','program_genre','original_name']].astype(str) # read string columns as string data types
```

```python
#Check the resolution type
resolution_counts = dataframe['hd'].value_counts()

# Display the counts
print(resolution_counts)
```

```python
# we import Visualization libraries
# you can ignore and use any other graphing libraries
import matplotlib.pyplot as plt # a comprehensive library for creating static, animated, and interactive visualizations
import plotly #a graphing library makes interactive, publication-quality graphs. Examples of how to make line plots, scatter plots, area charts, bar charts, error bars, box plots, histograms, heatmaps, subplots, multiple-axes, polar charts, and bubble charts.
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
# Calculate the percentage of views for each resolution
total_views = resolution_counts.sum()
resolution_percentages = (resolution_counts / total_views) * 100

# Create a DataFrame for Plotly
resolution_data = pd.DataFrame({
    'Resolution': ['SD', 'HD'],
    'Percentage': [resolution_percentages.get(0, 0), resolution_percentages.get(1, 0)]
})

# Plotting the pie chart using Plotly Express
fig = px.pie(
    resolution_data,
    names='Resolution',
    values='Percentage',
    title='Percentage of Views by Video Resolution',
    color='Resolution',
    color_discrete_sequence=['#2e8b57', '#4682b4'],  # Customize colors as needed
    labels={'Percentage': 'Percentage of Views'}
)

# Display the plot
fig.update_traces(textinfo='percent+label')  # Show percentage and label in the pie slices
fig.show()
```

## Conclusion:

**This visualization shows that the majority of video views were in SD resolution, accounting for 61.4% of total views. Meanwhile, views in HD resolution accounted for 38.6% of the total. This may indicate that users prefer watching videos in SD resolution more than HD, possibly due to preferences for video quality or limitations related to internet speeds or devices used.**

```python
# Create a copy of the original DataFrame to avoid modifying the original data
grouped = df.copy()

# Group the data by 'hd' and 'program_class'
# Aggregate the following:
# 1. The unique count of users who watched (nunique on user_id_maped)
# 2. The total count of watches (count on user_id_maped)
# 3. The sum of watch durations (sum on duration_seconds)
grouped = grouped.groupby(['hd', 'program_class']).agg({
    'user_id_maped': [('No of Users who Watched', 'nunique'),  # Unique users count
                      ('No of watches', 'count')],  # Total watch count
    'duration_seconds': [('Total watch time in seconds', 'sum')]  # Sum of watch duration in seconds
}).reset_index()

# Rename columns to be more readable after aggregation
grouped.columns = ['hd', 'program_class', 'No of Users who Watched', 'No of watches', 'Total watch time in seconds']

# Add a new column to convert total watch time from seconds to hours
grouped['Total watch time in hours'] = grouped['Total watch time in seconds'] / 3600

# Drop the "Total watch time in seconds" column as it is no longer needed
grouped = grouped.drop(columns=['Total watch time in seconds'])

# Sort the data by total watch time in hours, then by the number of watches, and finally by the number of users
grouped = grouped.sort_values(by=['Total watch time in hours', 'No of watches', 'No of Users who Watched'],
                              ascending=False).reset_index(drop=True)
```

```python
# show the result
grouped.head()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create subplots: 1 row and 2 columns, both of 'domain' type (for pie charts)
fig = make_subplots(1, 2, specs=[[{'type': 'domain'}, {'type': 'domain'}]],
                    subplot_titles=['SERIES/EPISODES', 'MOVIE'])

# Add pie chart for SERIES/EPISODES
fig.add_trace(go.Pie(labels=grouped[grouped['program_class'] == 'SERIES/EPISODES']['hd'],
                     values=grouped[grouped['program_class'] == 'SERIES/EPISODES']['No of Users who Watched'],
                     name='SERIES/EPISODES', direction='clockwise', sort=True),
              row=1, col=1)

# Add pie chart for MOVIE
fig.add_trace(go.Pie(labels=grouped[grouped['program_class'] == 'MOVIE']['hd'],
                     values=grouped[grouped['program_class'] == 'MOVIE']['No of Users who Watched'],
                     name='MOVIE', direction='clockwise', sort=True),
              row=1, col=2)

# Update layout with a title
fig.update_layout(title_text='Number of Users who Watched Vs Program Quality Flag (HD)')

# Show the plot
fig.show()
```

```python
import plotly.express as px

# Create a new column combining 'hd' and 'program_class' for clearer labels
# Convert 'hd' to string to avoid TypeError
grouped['hd_program_class'] = grouped['program_class'] + ' (' + grouped['hd'].astype(str) + ')'

# Define a color palette for a more professional look
colors = ['#003f5c', '#58508d', '#bc5090', '#ff6361', '#ffa600']  # Professional color palette

# Create pie chart using the new combined column for names
fig_hd_program = px.pie(grouped,
                        values='No of Users who Watched',
                        names='hd_program_class',
                        hover_data=['program_class'],
                        title='HD vs SD by Program Class: Number of Users',
                        color_discrete_sequence=colors)  # Apply the professional color palette

# Update the traces to show percentage and label inside the pie chart
fig_hd_program.update_traces(textinfo='percent+label',
                             textposition='inside',
                             showlegend=True)

# Customize layout for better readability
fig_hd_program.update_layout(
    title_font_size=18,
    title_x=0.5,  # Center the title
    annotations=[dict(text='HD/SD by Program Class', x=0.5, y=-0.1, showarrow=False)],
    legend_title="Program Class (HD/SD)",
    legend=dict(
        x=1,  # Position the legend on the right side
        y=1,
        bgcolor='rgba(255, 255, 255, 0.5)',
        bordercolor='Black',
        borderwidth=1
    )
)

# Show the updated pie chart
fig_hd_program.show()
```

## How does the average watch time for content in Standard Definition (SD) compare to that for High Definition (HD) resolution, and what might be the factors contributing to the observed differences in viewing duration?


```python

# Group the data by resolution (HD or SD) and calculate average watch time
grouped_by_resolution = dataframe.groupby('hd').agg({
    'duration_seconds': ['mean']
}).reset_index()

# Rename columns for clarity
grouped_by_resolution.columns = ['Resolution', 'Average Watch Time (Seconds)']

# Map resolution values to more readable labels
resolution_labels = {0: 'SD', 1: 'HD'}
grouped_by_resolution['Resolution'] = grouped_by_resolution['Resolution'].map(resolution_labels)

# Plot the average watch time by resolution
fig_average_watch_time = px.bar(grouped_by_resolution, x='Resolution', y='Average Watch Time (Seconds)',
                                title='Average Watch Time by Video Resolution',
                                labels={'Average Watch Time (Seconds)': 'Average Watch Time (Seconds)'},
                                color='Resolution',
                                color_discrete_sequence=['#1f77b4', '#ff7f0e'])  # Using more serious colors

# Display the plot
fig_average_watch_time.show()
```
