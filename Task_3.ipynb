## **TO DO build your Recommender system to Highlight Programs that users might be interested in**





```python
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors

# Ensure no duplicate values; if duplicates exist, compute the average rating
cleaned_dataframe = df.groupby(['program_name', 'user_id_maped'], as_index=False)['rating'].mean()

# Pivot the dataframe to create a user-item matrix with ratings
user_item_matrix = cleaned_dataframe.pivot(index='program_name', columns='user_id_maped', values='rating').fillna(0)

# Convert the DataFrame to a sparse matrix to save memory and improve performance
sparse_user_item_matrix = csr_matrix(user_item_matrix.values)

# Initialize and fit the NearestNeighbors model using cosine distance
knn_model = NearestNeighbors(metric='cosine', algorithm='brute')
knn_model.fit(sparse_user_item_matrix)

# Display the first few rows of the pivot table to inspect the matrix
print(user_item_matrix.head())
```

## **TO DO show the recommendations (top 5) for the people who watched "Moana" movie**

```python
# Data cleaning: Handle duplicates if necessary
# Calculate the average rating if there are duplicate entries
cleaned_df = df.groupby(['program_name', 'user_id_maped'], as_index=False)['rating'].mean()

# Create a user-item matrix with ratings
# Each row is a program, each column is a user, and values are ratings
user_item_matrix = cleaned_df.pivot(index='program_name', columns='user_id_maped', values='rating').fillna(0)

# Convert the DataFrame to a sparse matrix to save memory and improve performance
user_item_matrix_sparse = csr_matrix(user_item_matrix.values)

# Initialize and fit the NearestNeighbors model using cosine distance
knn_model = NearestNeighbors(metric='cosine', algorithm='brute')
knn_model.fit(user_item_matrix_sparse)

# Function to get similar items based on a given program
def get_similar_items(program_name, num_recommendations=5):
    # Find the index of the specified program
    if program_name not in user_item_matrix.index:
        raise ValueError(f"Program '{program_name}' not found in the dataset.")

    program_index = user_item_matrix.index.get_loc(program_name)

    # Find similar items using the NearestNeighbors model
    distances, indices = knn_model.kneighbors(user_item_matrix_sparse[program_index], n_neighbors=num_recommendations + 1)

    # Create a list of similar items
    similar_items = []
    for i in range(1, len(indices.flatten())):  # Start from 1 to avoid the item itself
        similar_item_index = indices.flatten()[i]
        similar_item = user_item_matrix.index[similar_item_index]
        similar_items.append((similar_item, distances.flatten()[i]))

    return similar_items

# Example of getting recommendations for a specific program
example_program = 'Moana'  # Replace with an program name from your data
recommendations = get_similar_items(example_program, num_recommendations=5)

print(f"Recommendations for '{example_program}':")
for item, distance in recommendations:
    print(f"Item: {item}, Distance: {distance:.2f}")
```
